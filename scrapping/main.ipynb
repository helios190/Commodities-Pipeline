```python
from jfx_scrapping import scrape_table_data
from BI_scrapping import download_inflation_data
from BPS_scrapping import fetch_and_transform_bps_data
import yfinance as yf
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date, col, expr, date_format
from pyspark.sql.types import DateType

# Dictionary to map Indonesian months to English months
indonesian_to_english_months = {
    'Januari': 'January', 'Februari': 'February', 'Maret': 'March', 'April': 'April',
    'Mei': 'May', 'Juni': 'June', 'Juli': 'July', 'Agustus': 'August',
    'September': 'September', 'Oktober': 'October', 'November': 'November', 'Desember': 'December'
}

# Spark function to replace Indonesian month names with English in a DataFrame column
def translate_months(df, column):
    for indo_month, eng_month in indonesian_to_english_months.items():
        df = df.withColumn(column, expr(f"regexp_replace({column}, '{indo_month}', '{eng_month}')"))
    return df

def normalize_date_formats(spark_df, date_column, current_format):
    # Convert date column to Spark DateType using the specified format and truncate to the first of each month
    return spark_df.withColumn("Date", to_date(col(date_column), current_format)).withColumn("Date", date_format("Date", "yyyy-MM-01").cast(DateType()))


spark = SparkSession.builder.appName("Data Collection").getOrCreate(
# JFX Data
urls = [
    'https://www.jfx.co.id/media?hal=hm-overview-harga&id=CC5'
    # 'https://www.jfx.co.id/media?hal=hm-overview-harga&id=GOL',
    # 'https://www.jfx.co.id/media?hal=hm-overview-harga&id=ACF',
    # 'https://www.jfx.co.id/media?hal=hm-overview-harga&id=RCF',
    # 'https://www.jfx.co.id/media?hal=hm-overview-harga&id=OLE'
]
start_date = '01-01-2023'
end_date = '01-01-2024'
df_jfx = scrape_table_data(urls, start_date, end_date)
spark_df_jfx = spark.createDataFrame(df_jfx)
spark_df_jfx = normalize_date_formats(spark_df_jfx, "BUSINESS DATE", "dd-MM-yyyy"
# BI Data
start_date = '01/2023'
end_date = '01/2024'
df_BI = download_inflation_data(start_date, end_date)
spark_df_BI = spark.createDataFrame(df_BI
# Translate months from Indonesian to English for the 'Periode' column in BI Data
spark_df_BI = translate_months(spark_df_BI, "Periode")
spark_df_BI = normalize_date_formats(spark_df_BI, "Periode", "MMMM yyyy")
spark_df_BI = spark_df_BI.drop("Periode"
# BPS Data
df_result = fetch_and_transform_bps_data()
spark_df_bps = spark.createDataFrame(df_result
# Translate months from Indonesian to English in BPS Data 'Date' column and normalize
spark_df_bps = translate_months(spark_df_bps, "Date")
spark_df_bps = normalize_date_formats(spark_df_bps, "Date", "yyyy MMMM"
# Yahoo Finance Data
ticker = 'IDR=X'
data = yf.download(ticker, start="2023-01-01", end="2024-01-01", interval='1d')
spark_df_yfinance = spark.createDataFrame(data.reset_index())
spark_df_yfinance = spark_df_yfinance.withColumnRenamed("Date", "yfinance_date"
# Truncate Yahoo Finance dates to monthly
spark_df_yfinance = spark_df_yfinance.withColumn("Date", date_format("yfinance_date", "yyyy-MM-01").cast(DateType()))
spark_df_yfinance = spark_df_yfinance.drop("yfinance_date"
# Perform outer joins on the normalized "Date" column
joined_data = spark_df_jfx \
    .join(spark_df_yfinance, "Date", "outer") \
    .join(spark_df_BI, "Date", "outer") \
    .join(spark_df_bps, "Date", "outer"
# Show the final combined data
joined_data.show(
# Convert to pandas if needed
final_result_df = joined_data.toPandas(
# Stop Spark session
spark.stop()



# Usage
df_final = collect_data()
print(df_final)
```

```python
import pandas as pd
import yfinance as yf

# Sample data for demonstration (replace with actual data fetching functions)

# Load JFX data
df_jfx = pd.DataFrame({
    'BUSINESS DATE': ['11-11-2024', '01-11-2024', '01-11-2024'],
    'CONTRACT': ['CC5', 'OLE', 'OLE'],
    'OPEN': [0, 16420, 16430],
    # other columns...
})

# Load BI data
df_BI = pd.DataFrame({
    'Periode': ['September 2024', 'Agustus 2024', 'Juli 2024'],
    'Data Inflasi': ['1.84 %', '2.12 %', '2.13 %']
})

# Load BPS data
df_result = pd.DataFrame({
    'Date': ['2009 Januari', '2009 Februari', '2024 Mei', '2024 Juni'],
    'BI_Rate': [8.75, 8.25, 6.25, 6.25]
})

# Dictionary to map Indonesian months to English months
indonesian_to_english_months = {
    'Januari': 'January', 'Februari': 'February', 'Maret': 'March', 'April': 'April',
    'Mei': 'May', 'Juni': 'June', 'Juli': 'July', 'Agustus': 'August',
    'September': 'September', 'Oktober': 'October', 'November': 'November', 'Desember': 'December'
}

# Function to replace Indonesian month names with English
def translate_months(date_str):
    for indo_month, eng_month in indonesian_to_english_months.items():
        if indo_month in date_str:
            return date_str.replace(indo_month, eng_month)
    return date_str

# Apply translation function to the date columns
df_BI['Periode'] = df_BI['Periode'].apply(translate_months)
df_result['Date'] = df_result['Date'].apply(translate_months)

# Convert the translated columns to datetime and truncate to month
df_jfx['Date'] = pd.to_datetime(df_jfx['BUSINESS DATE'], format='%d-%m-%Y').dt.to_period('M').dt.to_timestamp()
df_BI['Date'] = pd.to_datetime(df_BI['Periode'], format='%B %Y').dt.to_period('M').dt.to_timestamp()
df_result['Date'] = pd.to_datetime(df_result['Date'], format='%Y %B').dt.to_period('M').dt.to_timestamp()

# Drop the original date columns if they are not needed
df_jfx.drop(columns=['BUSINESS DATE'], inplace=True)
df_BI.drop(columns=['Periode'], inplace=True)

# Load Yahoo Finance data
ticker = 'IDR=X'
data = yf.download(ticker, start="2023-01-01", end="2024-01-01", interval='1d')
data.reset_index(inplace=True)
data['Date'] = pd.to_datetime(data['Date']).dt.to_period('M').dt.to_timestamp()

# Merge DataFrames on 'Date' using outer join to retain all months
merged_df = pd.merge(df_jfx, data, on='Date', how='outer')
merged_df = pd.merge(merged_df, df_BI, on='Date', how='outer')
merged_df = pd.merge(merged_df, df_result, on='Date', how='outer')

# Display the merged DataFrame
print(merged_df)
```
